Instructions

Create a new python file and import your Dog class from the previous exercise.

In the new python file, create a class named PetDog that inherits from Dog.

Add an attribute called trained to the __init__ method, this attribute is a boolean and the value should be False by default.

Add the following methods:

train: prints the output of bark and switches the trained boolean to True

play: takes a parameter which value is a few names of other Dog instances (use *args). The method should print the following string: ‚Äúdog_names all play together‚Äù.

do_a_trick: If the dog is trained the method should print one of the following sentences at random:
‚Äúdog_name does a barrel roll‚Äù.
‚Äúdog_name stands on his back legs‚Äù.
‚Äúdog_name shakes your hand‚Äù.
‚Äúdog_name plays dead‚Äù.


```python
import random
from Week3Day2ExercisesXPFile1 import Dog

class PetDog(Dog):
    def __init__(self, name, age, weight):
        super().__init__(name, age, weight)
        self.trained = False

    def train(self):
        print(self.bark())
        self.trained = True

    def play(self, *other_dogs):
        dog_names = [self.name] + [dog.name for dog in other_dogs]
        print(f"{', '.join(dog_names)} all play together")

    def do_a_trick(self):
        if self.trained:
            tricks = [
                f"{self.name} does a barrel roll",
                f"{self.name} stands on his back legs",
                f"{self.name} shakes your hand",
                f"{self.name} plays dead"
            ]
            print(random.choice(tricks))
        else:
            print(f"{self.name} is not trained yet.")

# Example usage:
pet_dog1 = PetDog(name="Perro", age=2, weight=15)
pet_dog2 = PetDog(name="Chien", age=3, weight=20)
pet_dog3 = PetDog(name="Hund", age=4, weight=22)

pet_dog1.train()
pet_dog1.do_a_trick()
pet_dog2.do_a_trick()
pet_dog1.play(pet_dog2, pet_dog3)
```

Exercise 4 : Family
Instructions
Create a class called Family and implement the following attributes:

members: list of dictionaries
last_name : (string)

Implement the following methods:

born: adds a child to the members list (use **kwargs), don‚Äôt forget to print a message congratulating the family.
is_18: takes the name of a family member as a parameter and returns True if they are over 18 and False if not.
family_presentation: a method that prints the family‚Äôs last name and all the members‚Äô details.

Create an instance of the Family class, with the last name of your choice, and the below members. Then call all the methods you created in Point 2.

    [
        {'name':'Michael','age':35,'gender':'Male','is_child':False},
        {'name':'Sarah','age':32,'gender':'Female','is_child':False}
    ]

```python
class Family:
    def __init__(self, last_name):
        self.members = []
        self.last_name = last_name

    def born(self, **kwargs):
        new_member = {'name': kwargs['name'], 'age': kwargs['age'], 'gender': kwargs['gender'], 'is_child': True}
        self.members.append(new_member)
        print(f"Congratulations, {kwargs['name']} is born!")

    def is_18(self, name):
        for member in self.members:
            if member['name'] == name and member['age'] >= 18:
                return True
        return False

    def family_presentation(self):
        print(f"Welcome to the {self.last_name} family!")
        for member in self.members:
            if member['is_child']:
                print(f"{member['name']} (child) - {member['age']} years old")
            else:
                print(f"{member['name']} - {member['age']} years old")
        print(f"We have {len(self.members)} members in our family.")

# Creates an instance of the Family class
family = Family('Doe')

# Adds members to the family
family.members = [
    {'name':'Michael','age':35,'gender':'Male','is_child':False},
    {'name':'Sarah','age':32,'gender':'Female','is_child':False}
]

# Adds a child to the family
family.born(name='John', age=0, gender='Male')

# Checks if a family member is over 18
print(f"Is Michael over 18? {family.is_18('Michael')}")
print(f"Is Sarah over 18? {family.is_18('Sarah')}")

# Prints the family presentation
family.family_presentation()
```

Exercise 5 : TheIncredibles Family
Instructions
Create a class called TheIncredibles. This class should inherit from the Family class:
This is no random family they are an incredible family, therefore the members attributes, will be a list of dictionaries containing the additional keys : power and incredible_name. (See Point 4)


Add a method called use_power, this method should print the power of a member only if they are over 18 years old. If not raise an exception (look up exceptions) which stated they are not over 18 years old.


Add a method called incredible_presentation which :

Print a sentence like ‚Äú*Here is our powerful family **‚Äù
Prints the family‚Äôs last name and all the members‚Äô details (ie. use the super() function, to call the family_presentation method)


Create an instance of the Incredibles class, with the ‚ÄúIncredibles‚Äù last name, and the below members.

    [
        {'name':'Michael','age':35,'gender':'Male','is_child':False,'power': 'fly','incredible_name':'MikeFly'},
        {'name':'Sarah','age':32,'gender':'Female','is_child':False,'power': 'read minds','incredible_name':'SuperWoman'}
    ]


Call the incredible_presentation method.


Use the born method inherited from the Family class to add Baby Jack with the following power: ‚ÄúUnknown Power‚Äù.


Call the incredible_presentation method again.

```python
class TheIncredibles(Family):
    def __init__(self, last_name):
        super().__init__(last_name)

    def use_power(self, name):
        for member in self.members:
            if member['name'] == name and member['age'] >= 18:
                print(f"{member['name']} uses their power: {member['power']}")
            else:
                raise Exception(f"{name} is not old enough to use their power. They need to be at least 18 years old.")

    def incredible_presentation(self):
        print("*Here is our powerful family **")
        super().family_presentation()
        print("We are a family of superheroes, and we are here to save the day!")

# Create an instance of the TheIncredibles class
incredibles = TheIncredibles("Incredibles")

# Add members to the family
incredibles.members = [
    {'name':'Michael','age':35,'gender':'Male','is_child':False,'power': 'fly','incredible_name':'MikeFly'},
    {'name':'Sarah','age':32,'gender':'Female','is_child':False,'power': 'read minds','incredible_name':'SuperWoman'}
]

# Call the incredible_presentation method
incredibles.incredible_presentation()

# Use the born method to add Baby Jack with the power: ‚ÄúUnknown Power‚Äù
incredibles.born(name='Baby Jack', age=0, gender='Male', power='Unknown Power')

# Call the incredible_presentation method again
incredibles.incredible_presentation()

# Try to use Baby Jack's power
try:
    incredibles.use_power('Baby Jack')
except Exception as e:
    print(e)
```
