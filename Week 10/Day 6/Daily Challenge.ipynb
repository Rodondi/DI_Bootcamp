**1. Data Preparation**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# set seed for reproducibility
np.random.seed(42)

# Generate temperature data for 10 cities, 12 months
temps = np.random.uniform(low=-5, high=35, size=(10,12))

# Define cities and months
cities = [f"City_{i+1}" for i in range(10)]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Create the Dataframe
df_temps = pd.DataFrame(data=temps, index=cities, columns=months)
df_temps.head()
```

**2. Data Analysis**

```python
# Annual average temperature for each city
df_temps["Annual Avg"] = df_temps.mean(axis=1)

# Find city with highest and lowest average temperature
hottest_city = df_temps["Annual Avg"].idxmax()
coldest_city = df_temps["Annual Avg"].idxmin()

# Display results
print(f" Hottest City: {hottest_city} with {df_temps.loc[hottest_city, 'Annual Avg']:.2f}°C")
print(f" Coldest City: {coldest_city} with {df_temps.loc[coldest_city, 'Annual Avg']:.2f}°C")
```

**3. Data Visulaization**

**Line Plot: Monthly Temperature Trends by City**

```python
plt.figure(figsize=(14, 6))
for city in df_temps.index:
  plt.plot(months, df_temps.loc[city, months], label=city)
plt.title("Monthly Average Temperatures by City")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Bar Chart: Annual Average Temperature Comparison**

```python
df_temps["Annual Avg"].sort_values().plot(kind='barh', color ='skyblue', figsize=(10, 6))
plt.title("Annual Average Temperature per City")
plt.xlabel("Temperature (°C)")
plt.axvline(df_temps["Annual Avg"].mean(), color='red', linestyle='--', label='Overall Average')
plt.legend()
plt.tight_layout()
plt.show()
```
