**Importing Required Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

**Data Import and Cleaning**

```python
# Load the Dataset
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('Airplane_Crashes_and_Fatalities.csv', encoding='ISO-8859-1')
df
```

```python
df.info()
```

```python
# Renaming inconsistent columns
df.rename(columns={
    'Aboard Passangers': 'Aboard Passengers',
    'Fatalities Passangers': 'Fatalities Passengers'
}, inplace=True)
```

```python
# Convert Date and Time columns
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce').dt.time
```

```python
# Fill missing numeric calues with 0 for simplicity
df.fillna({
    'Aboard': 0,
    'Aboard Passengers':0,
    'Aboard Crew': 0,
    'Fatalities': 0,
    'Fatalities Passengers': 0,
    'Fatalities Crew': 0,
    'Ground': 0,
}, inplace=True)
df.info()
```

**Exploratory Data Analysis**

```python
print('Total Crashes:', len(df))
print('Total Fatilities:', df['Fatalities'].sum())
print('Average Fatalities per Crash:', df['Fatalities'].mean())
```

```python
# Crahes per Year
df['Year'] = df['Date'].dt.year
crashes_per_year = df.groupby('Year').size()
crashes_per_year
```

**Visualization**

```python
# Crashes Over Time
plt.figure(figsize=(12, 5))
crashes_per_year.plot()
plt.title('Airplane Crashes per Year')
plt.xlabel('Year')
plt.ylabel('Number of Crashes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Histogram of Fatalities
plt.figure(figsize=(8,5))
sns.histplot(df['Fatalities'], bins=50, kde=True)
plt.title('Distribution of Fatalities per Crash')
plt.xlabel('Fatalities')
plt.ylabel('Number of Crashes')
plt.tight_layout()
plt.show()
```

**Statistical Analysis**

```python
fatal_mean = df['Fatalities'].mean()
fatal_median = df['Fatalities'].median()
fatal_std = df['Fatalities'].std()

print('Mean Fatalities:', fatal_mean)
print('Median Fatalities:', fatal_median)
print('Standard Deviation:', fatal_std)
```

```python
# Hypothesis Test: Compare average fatalities in two decades
df['Decade'] = (df['Year'] // 10) * 10
decade_group = df[df['Year'] >= 1950].groupby('Decade')['Fatalities']
decades = list(decade_group.groups.keys())
if len(decades) >= 2:
    group1 = decade_group.get_group(decades[0])
    group2 = decade_group.get_group(decades[1])
    t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=False)
    print(f"T-test between {decades[0]}s and {decades[1]}s: t = {t_stat:.2f}, p = {p_val:.4f}")
```

```python
# Top 10 Operators by Crash Account
plt.figure(figsize=(10, 6))
df['Operator'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Operators by Number of Crashes')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
