```python
df = pd.read_csv("ds_salaries.csv", index_col=0)
df.head()
```

**Identify Quantitative vs Qualitative Data**

**Quantitative: Numeric/measurable values**

salary, salary_in_usd

**Qualitative (Categorical): Descriptive values**

experience_level, employment_type, job_title, company_location, company_size.

**Check and Handle Missing Values**

```python
df.isnull().sum()
```

**Group-wise Analysis by Experience Level**

```python
grouped = df.groupby('experience_level')['salary_in_usd'].agg(['mean', 'median']).reset_index()
grouped
```

**Bar Chart of Average Salary by Experience Level**

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.bar(grouped['experience_level'], grouped['mean'])

# Labels and Title
plt.xlabel('Experience Level')
plt.ylabel('Average Salary(USD)')
plt.title('Average Data Science Salary by Experience Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```
