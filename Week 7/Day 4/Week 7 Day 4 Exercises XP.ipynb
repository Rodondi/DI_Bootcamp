**Exercise 1: Introduction to Data Analysis**
**What is Data Analysis?**

Data analysis is the systematic process of collecting, organizing, interpreting, and presenting data in order to extract useful information, identify patterns, and support decision-making.

**Why is data analysis important in modern contexts?**

Data analysis helps organizations in the curent data-driven world to:

*   Drive strategic business decisions
*   Identify and forecast trends
*   Improve operational efficiency
*   Detect anomalies and reduce risks

**Three Areas Where Data Analysis is Applied**

**Healthcare:**

Data analysis is used to monitor patient outcomes, predict disease outbreaks, and optimize hospital resources. For example, analyzing patient data can help detect early signs of chronic conditions.

**Finance:**

In the financial sector, data analysis is used for credit scoring, fraud detection, and investment forecasting. Banks and fintech firms rely on data models to assess risk and personalize financial products.

**Agriculture:**

Farmers and agribusinesses use data analysis to track crop performance, monitor soil health, and optimize irrigation and fertilizer usage. This helps improve yield and sustainability.

## Exercise 3: Identifying Data Types

### Credit Card Approvals Dataset

* **Gender** – Qualitative: It shows if the person is male or female.
* **Age** – Quantitative: It’s a number that tells how old someone is.
* **Debt** – Quantitative: A number showing how much debt someone has.
* **Married** – Qualitative: Tells whether someone is married or not.
* **BankCustomer** – Qualitative: Tells whether someone is a customer of the bank.
* **Industry** – Qualitative: Describes the type of job or work sector.
* **Ethnicity** – Qualitative: Describes a person’s ethnic group.
* **YearsEmployed** – Quantitative: Tells how many years the person has been working.
* **PriorDefault** – Qualitative: Tells if the person had a default before.
* **Employed** – Qualitative: Tells if the person is currently employed.
* **CreditScore** – Quantitative: A number that shows how good someone’s credit is.
* **DriversLicense** – Qualitative: Tells if the person has a driver’s license.
* **Citizen** – Qualitative: Tells the person’s citizenship status (given as a category).
* **ZipCode** – Qualitative: A postal code – even though it's a number, it represents a place, not a quantity.
* **Income** – Quantitative: A number showing how much income someone earns.
* **Approved** – Qualitative: Tells if the credit card application was approved or not.

### Mental Health Disorder Dataset

* **Entity** – Qualitative: The name of the country or region.
* **Code** – Qualitative: A short code that stands for the country.
* **Year** – Quantitative: The year the data was recorded.
* **Schizophrenia (%)** – Quantitative: A percentage showing how many people had schizophrenia.
* **Bipolar disorder (%)** – Quantitative: A percentage showing how many people had bipolar disorder.
* **Eating disorders (%)** – Quantitative: A percentage showing how many people had eating disorders.
* **Anxiety disorders (%)** – Quantitative: A percentage showing how many people had anxiety.
* **Drug use disorders (%)** – Quantitative: A percentage showing how many people had drug problems.
* **Depression (%)** – Quantitative: A percentage showing how many people were depressed.
* **Alcohol use disorders (%)** – Quantitative: A percentage showing how many people had alcohol-related issues.

### Time Americans Spend Sleeping Dataset

* **Year** – Quantitative: The year the data was recorded.
* **Period** – Qualitative: A time range like “2005–2009.”
* **Avg hrs per day sleeping** – Quantitative: The average number of hours people slept.
* **Standard Error** – Quantitative: A small number that shows how much the average might vary.
* **Type of Days** – Qualitative: Describes whether it was a weekday or weekend.
* **Age Group** – Qualitative: Groups people by their age range.
* **Activity** – Qualitative: Describes what the person was doing – in this case, it's always “Sleeping.”

Exercise 5: Basic Data Analysis with Google Colab

```python
# Mean
mean_sepal_length = iris['SepalLengthCm'].mean()

# Median
median_sepal_length = iris['SepalLengthCm'].median()

# Mode
mode_sepal_length = iris['SepalLengthCm'].mode()[0]

print("Mean Sepal Length:", mean_sepal_length)
print("Median Sepal Length:", median_sepal_length)
print("Mode Sepal Length:", mode_sepal_length)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Sepal Length
plt.figure(figsize=(8, 5))
sns.histplot(iris['SepalLengthCm'], bins=15, kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Statistical Measures (SepalLengthCm)

- **Mean**: The average sepal length is **5.84 cm**
- **Median**: The middle value of sepal length is **5.8 cm**
- **Mode**: The most common sepal length is **5.0 cm**

### Visualization: Histogram of Sepal Length

- Most sepal lengths fall between **5.0 cm and 6.5 cm**
- The distribution is slightly skewed to the right, meaning there are a few longer sepals
- The **KDE curve** (smooth line) helps visualize the shape of the distribution
