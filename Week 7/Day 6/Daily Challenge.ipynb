To load an image and display it :

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load image
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display original image
plt.imshow(original_image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

Rotate an image by 90 degrees

```python
from scipy.ndimage import rotate
import numpy as np

def rotate_image_90_degrees(image):
    image_np = np.array(image)
    rotated = rotate(image_np, 90, reshape=True)
    return Image.fromarray(rotated.astype('uint8'))

rotated_image = rotate_image_90_degrees(original_image)
plt.imshow(rotated_image)
plt.title("Rotated by 90°")
plt.axis('off')
plt.show()
```

Flip an image horizontally and then vertically using mirror and flip functions from ImageOps

```python
from PIL import ImageOps

# Flip Horizomntally
flipped_horizontal = ImageOps.mirror(original_image)

#Flip Vertically
flipped_vertical = ImageOps.flip(original_image)

# Displaying the flipped images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(flipped_horizontal)
plt.title("Flipped Horizontally")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(flipped_vertical)
plt.title("Flipped Vertically")
plt.axis('off')

plt.show()
```

Zoom in on an image (scale by 1.2x) using .resize(...)

```python
def zoom_image(image, zoom_factor=1.2):
  width, height = image.size
  new_width = int(width * zoom_factor)
  new_height = int(height * zoom_factor)
  return image.resize((new_width, new_height))

zoomed_image = zoom_image(original_image)
plt.imshow(zoomed_image)
plt.title("Zoomed In (1.2x)")
plt.axis('off')
plt.show()
```
